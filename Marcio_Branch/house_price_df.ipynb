```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns
import matplotlib.cm as cm

# File to Load
dt_dc_prices_df= 'Datasets/dc_housing_all_clean-2008-2016.csv'

# Read crimes Data File and store into Pandas Data Frames
dc_prices = pd.read_csv(dt_dc_prices_df)

dc_prices.describe()
```

```python
dc_prices.Median_Price.describe()
```

```python
# Generate a bar plot showing number of meadian price for each cluster using pandas
sum_house_cluster1 = dc_prices.groupby(['Cluster'])['Median_Price'].median()

median_house_graph = sum_house_cluster1.plot(kind = 'bar', figsize=(12,8))

plt.title('Median Price by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Median Price')
```

```python
#Generate a bar plot showing number of data points for each cluster using pandas
sum_house_cluster2 = dc_prices.groupby(['Cluster'])['Median_Price'].sum()#.sort_index()

sum_median_house = sum_house_cluster2.plot(kind='bar', figsize=(12,8))

#ax = sum_house_cluster2.iloc[:41].plot(kind='bar', figsize=(12,8))
#ax.invert_yaxis()

plt.title('Sum of Median Prices by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Sum of Median Price')
```

```python
#Generate a scatter plot crossing Clusters, Median Price by Year

sum_house_cluster3 = dc_prices.groupby(['Cluster', 'Year'])['Median_Price'].median().reset_index()

 #applies the custom color map along with the color sequence

plt.figure(figsize=(20,30))
plt.scatter(x=sum_house_cluster3['Median_Price'], y=sum_house_cluster3['Median_Price'], alpha=0.70, s=300, c= sum_house_cluster3['Year'], cmap=cm.brg)


plt.title('Clusters vs Median Price by Year', fontsize=14)
plt.xlabel('Median Price')
plt.ylabel('Cluster')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
```

```python
#Generate a scatter plot crossing Clusters, Variance of the Median Price over Year

sum_house_cluster3_1 = dc_prices.groupby(['Cluster'])['Median_Price'].var()

var_median_house = sum_house_cluster3_1.plot(kind='bar', figsize=(12,8))
```

```python
# Generate a line plot

sum_house_cluster4 = dc_prices.groupby(['Year'])['Median_Price'].median().reset_index()

ax=sum_house_cluster4.plot(x='Year', y='Median_Price',figsize=(12,8))
ax.set_title('Median Price by Year', fontsize=14)
ax.set_ylabel('Median Price')
```
